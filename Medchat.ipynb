# 2.**试试基础模型**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 指定模型名称
model_name = "BioMistral/BioMistral-7B"

# 加载模型和分词器
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def chat_with_model(question):
    """ 使用模型回答医学问题 """
    inputs = tokenizer(question, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# 测试模型
question = "What are the symptoms of pneumonia?"
response = chat_with_model(question)
print("BioMistral Answer:", response)
```

```python
def chat_with_model(question):
    """ 让医学模型回答问题 """
    inputs = tokenizer(question, return_tensors="pt")

    outputs = model.generate(
        **inputs,
        max_length=100,
        temperature=0.7,  # 控制生成的随机性
        top_p=0.9,  # 限制采样范围
        do_sample=True  # 让回答更自然
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# 测试改进后的生成方法
question = "What are the symptoms of having a fever?"
response = chat_with_model(question)
print("Improved Model Answer:", response)
```

```python
import json
from datasets import Dataset

# 读取 JSON 数据
with open("ori_pqal.json", "r") as f:
    raw_data = json.load(f)

# 提取 QA 数据
qa_data = {"question": [], "context": [], "answer": []}
for key, value in raw_data.items():
    qa_data["question"].append(value["QUESTION"])
    qa_data["context"].append(" ".join(value["CONTEXTS"]))
    qa_data["answer"].append(value["LONG_ANSWER"])

# 转换为 Hugging Face 数据集
hf_dataset = Dataset.from_dict(qa_data)

# 确保数据集正确加载
print("🔍 解析完成，数据示例：", hf_dataset[0])


```

```python
# Tokenization 预处理
def tokenize_data(example):
    input_text = f"Question: {example['question']} Context: {example['context']} Answer:"
    target_text = example["answer"]

    input_encoding = tokenizer(input_text, padding="max_length", truncation=True, max_length=512)
    target_encoding = tokenizer(target_text, padding="max_length", truncation=True, max_length=512)

    labels = target_encoding["input_ids"]
    labels = [-100 if token == tokenizer.pad_token_id else token for token in labels]  # 避免计算 loss

    return {
        "input_ids": input_encoding["input_ids"],
        "attention_mask": input_encoding["attention_mask"],
        "labels": labels
    }

# 重新 Tokenize 数据集
tokenized_dataset = hf_dataset.map(tokenize_data)

# 确保转换正确
print("🔍 Tokenized 数据示例：", tokenized_dataset[0])

```

```python
from transformers import AutoModelForCausalLM
from peft import LoraConfig, get_peft_model

# 加载 BioGPT 预训练模型
model_name = "microsoft/BioGPT"
model = AutoModelForCausalLM.from_pretrained(model_name)

# 确保 BioGPT 被 PEFT 识别
model.config.model_type = "biogpt"

# BioGPT 的 LoRA 目标层
target_modules = ["self_attn.q_proj", "self_attn.k_proj", "self_attn.v_proj", "self_attn.out_proj", "fc1", "fc2"]

# 定义 LoRA 适配器
lora_config = LoraConfig(
    r=8, lora_alpha=16, lora_dropout=0.1,
    task_type="CAUSAL_LM",
    target_modules=target_modules
)

# 应用 LoRA
model = get_peft_model(model, lora_config)

print("✅ LoRA 适配 BioGPT 成功！🚀")

```

```python
from transformers import TrainingArguments, Trainer

# 训练参数
training_args = TrainingArguments(
    per_device_train_batch_size=3,
    num_train_epochs=3,
    logging_dir="./logs",
    output_dir="./finetuned_biogpt",
    save_strategy="epoch"
)


small_dataset = tokenized_dataset.select(range(1000))
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_dataset
)
trainer.train()



# 保存微调后的模型
model.save_pretrained("./finetuned_biogpt")
tokenizer.save_pretrained("./finetuned_biogpt")

print("🎉 LoRA 微调 BioGPT 完成！🚀")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 加载微调后的 BioGPT
model_path = "./finetuned_biogpt"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)

def chat_with_bot(question, context):
    prompt = f"Question: {question} Context: {context} Answer:"
    inputs = tokenizer(prompt, return_tensors="pt")
    output = model.generate(**inputs, max_length=150)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# 试试你的微调模型
question = "What are the effects of mitochondria on programmed cell death?"
context = "Mitochondria play a crucial role in cell apoptosis and programmed cell death..."
response = chat_with_bot(question, context)
print("MedBot Answer:", response)

```
